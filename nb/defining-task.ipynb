Постановка задачи (дополняется)

### Постановка задачи

Цель - подбор "видовых" домов и квартир в Москве. Пожелания: полу- или автоматический сбор данных, возможность интерактивного подбора на карте, в идеале - веб-сервис, взаимодействующий с базами `cian.ru` или ещё кого.

### Сбор, предобработка данных
Этому посвящён отдельный ноутбук `data-preparation`, на текущий момент данные собираются руками из открытых источников.

### Качество вида - 1
Интуитивно предположим, что вид из дома тем лучше, чем он выше - никто не перекроет обзор с последних этажей. И, чем больше разница в высоте (этажности) с соседями, тем более видовым будет считаться дом. 

Таким образом, задача сводится к поиску локальных максимумов высоты (этажности) в заданном радиусе на координатной сетке. 

- Следует учесть рельеф земной поверхности. 
- Высота одного этажа всюду разная, в общем случае.

Необходимые данные:
- пары: местоположение, (высотность здания + высотность рельефа)

Местоположение может задаваться адресом, GPS, или своей собственной сеткой координат. Высотность известна для типовых и рекордных зданий, для остальных - нужно приближённо вычислять, зная их этажность. 

### Качество вида - 2
Предыдущая модель, увы, не учитывает особенности геометрии зданий и групп зданий. Здание может быть локально высоким, однако если оно имеет множество близких соседей (например, в составе ЖК), то вид может быть испорчен. 

Первая модель одинакова для любого из фасадов зданий (направлений обзора). К примеру,  для невысоких домов, расположенных между рекой и небоскрёбом, она может показать плохой результат, несмотря на открывающийся из них вид на реку.

Для формализации вышесказанного имеет смысл применить тригонометрию, трассировку лучей, и посчитать угол и дальность обзора для каждого фасада. Взяв, допустим, максимальный результат по всем фасадам, получим высшую возможную оценку для дома. Для простоты модели её можно вновь считать по максимальному этажу.

Необходимые данные:
- данные горизонтального сечения здания (его 'подошвы'), например, в виде полигонов
- данные высотности

Данную задачу можно решать как 'в горизонте', подсчитывая угол на плоскости, параллельной этажу, так и в объёме, определяя телесные углы обзора зрителя.

### Качество вида - 3
Заметим, что "крутость вида" - характеристика не только количественная (обзорность), но и качественная (предмет наблюдения). Качественная состовляющая, увы, субъективна - насколько хорош вид "одним глазком на Кремль" в сравнении с панорамой леса и холмов?

Таким образом, для конкретного зрителя можно ввести предпочтения объектов на карте, и строить рейтинг вида, основываясь на видимости этих объектов.
